```python
import numpy as np
import pandas as pd

#Site do dataset
url = 'https://raw.githubusercontent.com/mikecunha/pygeo_heatmap/master/data/campaign_contributions_for_map.tsv'

#Importando URL para DataFranme do Pandas
dataset = pd.read_csv(url, sep = '\t')
```

```python
#Apenas cidade do estado de Nova York
dataset_ny = dataset[dataset['State']=='NY']

dataset_ny
```

```python
#Dataframe de localização
gps = dataset_ny[['lat', 'lon']]

#Quantia por cidade
Amount = dataset_ny['Amount']

#Preparando o mapa
fig = gmaps.figure()

#Camada Heatmap
fig.add_layer(gmaps.heatmap_layer(gps, weights=Amount,
                                  max_intensity = float(dataset_ny['Amount'].mean()),
                                  point_radius = 7,
                                  opacity = 1))

#Exportando o gráfico para HTML
embed_minimal_html('heatmap.html', views = [fig])
```
